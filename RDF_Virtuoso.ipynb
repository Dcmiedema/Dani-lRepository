```python
class RDF_files_to_Virtuoso:
    
    count_files = 0
    
    def __init__(self, rdf_file, endpoint_url, username, password):
        self.endpoint_url = endpoint_url
        self.username = username
        self.password = password
        self.rdf_file = rdf_file

        
    def rdf_put(self):
        
        if isinstance(self.rdf_file,list):
            for i in self.rdf_file:
                with open(i, 'r') as working_file:
                    response = requests.put(self.endpoint_url + str(RDF_files_to_Virtuoso.count_files) + '.ttl',
                    data=working_file.read(),
                    auth=HTTPBasicAuth(self.username, self.password)
                )
                    working_file.close()
                    RDF_files_to_Virtuoso.count_files += 1

                    if response.status_code == 201:
                        print('succesful')
                    else:
                        print('failed',response)
            
        else:
            with open(self.rdf_file, 'r') as working_file:
                response = requests.put(self.endpoint_url + str(count_files) + '.ttl',
                data=working_file.read(),
                auth=HTTPBasicAuth(self.username, self.password)
            )
                working_file.close()
                RDF_files_to_Virtuoso.count_files += 1

                if response.status_code == 201:
                    print('succesful')
                else:
                    print('failed',response)
        
        return
    
```
